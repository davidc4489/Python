```python
import math, numpy, pandas
from sklearn.datasets import load_iris

def new_iris(new_iris,numneighbors):
    iris = load_iris()

    irisdata = iris["data"]

    lines = len(iris.data)

    distances = [0]*lines

    for i in range(lines):
        distances[i] = math.sqrt((irisdata[i, 0] - new_iris[0])**2 + (irisdata[i, 1] - new_iris[1])**2 + 
            (irisdata[i, 2] - new_iris[2])**2 + (irisdata[i, 3] - new_iris[3])**2)

    index = [0]*(numneighbors+1)
    for i in range(numneighbors+1):    
        minpos = distances.index(min(distances))
        index[i]=minpos
        distances[distances.index(min(distances))]= numpy.inf

    species = []
    for i in range(len(index)):
        species.append(iris["target"][index[i]])
    
    from collections import Counter
 
    occurence_count = Counter(species)
    specy =  occurence_count.most_common(1)[0][0]
    if len(occurence_count.most_common(1)) != 1:
        if occurence_count.most_common(1)[1][1]==occurence_count.most_common(1)[0][1]:
            return new_iris(new_iris,numneighbors-1)
    else:
        return specy

new_iris([7.7,2.6,6.9,2.3],5)


    
```
