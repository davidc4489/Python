```python
import numpy as np

def drops(n,h):
    if(n == 1 or h == 0 or h == 1):
        return h
    else:

        minimum = np.inf

        for i in range (1,h):
            minimum = min(minimum, 1 + max(drops(n - 1, i - 1), drops(n, h - i)))

        return minimum

drops(2,27)
```

```python
import numpy as np

def solvepuzzle(N,k):

    numdrops =  [(i,j) for i in range(1,N+1) for j in range(k+1)]
    print(numdrops)
    drops = {count:np.inf for count in numdrops}

    for i in range (0, N+1):
        drops[i,1] = 1
        drops[i,0] = 0
    print(drops)

    for i in range (0,k+1):
        drops[1, i] = i

 

    for i in range (2,N+1):
        for j in range (2,k+1):

            minimum = np.inf

            for x in range (1,j):
                minimum = min(minimum, 1 + max(drops[i-1,x-1],drops[i,j-x]))

            drops[i,j] = minimum


    return drops[N,k]

print(solvepuzzle(7,1000))

""" N = 2
k = 100
numdrops =  [(i,j) for i in range(N) for j in range(k)]
print(numdrops)
drops = {index:-1 for index in numdrops}
print(drops)
drops[0,2]=3
print(drops) """
```
