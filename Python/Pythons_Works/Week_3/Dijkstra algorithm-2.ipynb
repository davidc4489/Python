```python
w,h = img.shape
```

```python
def is_valid(x):
    i,j=x
    if i<w and i>=0 and j<h and j>=0:
        return True
```

```python
seen= set()
cost={}
parent ={}
cost[(5,25)] = 0
gg=(5,25)
hh=[]
while gg!=(220,5):
    if is_valid(gg)==True:
    
        seen.add(gg)
        rr=gg[0]+1,gg[1]
        ll=gg[0]-1,gg[1]
        oo=gg[0],gg[1]+1
        dd=gg[0],gg[1]-1
        aaa=[rr,ll,oo,dd]
        for i in aaa:
            if is_valid(i)==True:
                if i in seen:
                    continue
                elif img[i]>200:
                    if i in cost:
                        if cost [i]>cost[gg]+1:
                            parent[i] = gg
                            cost[i]=cost[gg]+1
                    else:
                        cost[i]=cost[gg]+1
                        parent[i] = gg
                    item = (cost[i],i)
                    heap.heappush(hh,item)
                
        gg=heap.heappop(hh)[1]
        while gg in seen:
            gg=heap.heappop(hh)[1]
    





allparen=[]
sss=gg
while sss!=(5,25):
    allparen.append(sss)
    sss=parent[allparen[-1]]





path = allparen
```

```python
def drawPath(img,path, thickness=1):
    '''path is a list of (x,y) tuples'''
    x0,y0=path[0]
    for vertex in path[1:]:
        x1,y1=vertex
        cv2.line(img,(y0,x0),(y1,x1),(255,0,0),thickness)
        x0,y0=vertex
```

```python
img = cv2.imread(file)
drawPath(img,path,2)
plt.figure(figsize=(7,7))
plt.imshow(img) # show the image on the screen 
plt.show()
```
