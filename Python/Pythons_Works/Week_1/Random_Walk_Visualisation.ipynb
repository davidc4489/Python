```python
# figure : une figure qui peut être sauvée dans un fichier.
# axe : un graphe (graphique) individuel appartenant à une figure qui peut en compter
# plusieurs. Attention, ce n'est pas un axe de coordonnées !
# axis : un axe de coordonnées appartenant à un objet axe.

# importation de la librairie qui va permettre de tracer le graphe 
# + appellation de la librairie sous le nom de plt
import matplotlib.pyplot as plt

# importation de la librairie qui va permettre des choix aléatoires
import random

# création d'une fonction pour retourner la liste des valeurs de x 
# + paramètres que reçoit la fonction
def random_walk_2_Directions (steps):
    # initialisation de la position de départ
    position = 0
    # création d'une liste avec les valeurs de x
    walk = [position]
    # boucle pour insérer les valeurs de x à la liste
    for i in range (steps):
        # changement de position aléatoire
        position += random.choice([-1,1])
        # ajout de la position à la liste
        walk.append(position)

    # la fonction retourne la liste des valeurs de x
    return walk

# création d'une fonction pour créer un graphe à partir de la liste des valeurs de x
# + paramètres que reçoit la fonction
def plot_random_walk (walk):

    # création d'un graphe + liste des valeurs de x comme paramètre de x 
    # + paramètre de y par défaut
    # syntaxe : plt.plot(X, Y, lw = épaisseur, ls = style, c = couleur)
    # lien vers tous les arguments possibles de plt.plot:
    # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
    plt.plot(walk)
    # titre du graphe
    plt.title("Random_Walk_1D")
    # légende de l'axe x
    plt.xlabel('position')
    # légende de l'axe y
    plt.ylabel('time')
    # impression du graphe à l'écran
    plt.show()

steps_for_2_directions = input("please enter number of steps")
plot_random_walk(random_walk_2_Directions(int(steps_for_2_directions)))
```

```python
# création d'une fonction pour retourner la liste des valeurs de x et de y 
# + paramètres que reçoit la fonction
def random_walk_4_Directions(steps):
    # initialisation des valeurs de x et de y
    x, y = 0, 0
    # création d'une liste avec les valeurs de x
    walk_x = [0]
    # création d'une liste avec les valeurs de y
    walk_y = [0]
    # boucle pour insérer les valeurs de x et de y à leur liste
    for i in range(steps):
        # changement de position aléatoire
        direction = random.choice(['North', 'South', 'West', 'East'])
        if direction == 'North':
            y +=1
        elif direction == 'South':
            y -=1
        elif direction == 'West':
            x -=1
        else:
            x +=1
        # ajout de la position à la liste
        walk_x.append(x)
        walk_y.append(y)
    # la fonction retourne les liste des valeurs de x et de y
    return walk_x, walk_y

# création d'une fonction pour créer un graphe à partir de la liste des valeurs de x et de y
# + paramètres que reçoit la fonction
def plot_random_walk_2(walk_x, walk_y):
    # création d'un graphe 
    # + liste des valeurs de x comme paramètre de x 
    # + liste des valeurs de y comme paramètre de y
    # Scatter plot (nuage de points)
    plt.scatter(walk_x, walk_y)
    # titre du graphe
    plt.title("Random_Walk_2D")
    # légende de l'axe x
    plt.xlabel('walk_x')
    # légende de l'axe y
    plt.ylabel('walk_y')
    # affichage d'une grille
    plt.grid()

    # impression du graphe à l'écran
    plt.show()

steps_for_4_directions = int(input("please enter number of steps"))
x, y = random_walk_4_Directions(steps_for_4_directions)
plot_random_walk_2(x,y)
```

```python
def random_walk_3d(walk_x, walk_y):
    time = list(range(len(walk_x)))
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.plot(walk_x, walk_y, time)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Time')

# Un graphique plotly est un objet Figure comprenant à minima deux attributs :
# - data : une séquence de traces contenant les données à afficher. Ex :
#   [Scatter(...), Bar(...)]
# - layout : un dictionnaire de paires clé/valeur déterminant titre, légende, axes,
# etc...
```

```python

# Pour alléger le code, on importe la sous bibliothèque nécessaire avec un alias
import plotly.graph_objects as go

# création d'une fonction pour créer un graphe à partir de la liste des valeurs de x et de y
# + paramètres que reçoit la fonction
def plot_random_walk_3d(walk_x, walk_y):
    steps = list(range(len(walk_x)))
    # On construit la trace avec l’objet Scatter
    # iI est possible d’utiliser la classe go.Scatter3D de plotly.graph_objects
    # Paramètres:
    # x – Définit les coordonnées x.
    # y – Définit les coordonnées y.
    # z – Définit les coordonnées z.
    # mode – Détermine le mode de dessin pour cette trace de dispersion.
    trace = go.Scatter3d( x = walk_x,
                          y = walk_y,
                          z = steps,
                          mode = 'lines',
                          line = dict(color = 'blue', width = 2))
    # Toujours conformément à la documentation, la propriété data doit être une séquence de traces.
    data = [trace]
    # L’aspect général du graphique est géré par l’objet Layout, pour lequel on précise:
    # - un Title ;
    # - un XAxis ;
    # - un YAxis ;
    layout = go.Layout (scene=dict(
                        xaxis_title='Y',
                        yaxis_title='X',
                        zaxis_title='Time',
                        ), title='Random Walk in 3D')
    # La figure est construite avec fig = go.Figure(data=data, layout=layout)
    fig = go.Figure(data=data, layout=layout)
    # impression du graphe à l'écran
    fig.show()

# interaction avec l'utilisateur 
# + réception du nombre de pas de l'utilisateur 
# + conversion des données en int
steps = int(input('Enter the number of steps: '))
# déclaration de 2 variables 
# + initialisation des variables à partir des listes que la fonction a retourné
walk_x, walk_y = random_walk_4_Directions(steps)
# utilisation de la fonction plot_random_walk_3d avec les paramètres reçus de l'utilisateur
plot_random_walk_3d(walk_x, walk_y)
```
