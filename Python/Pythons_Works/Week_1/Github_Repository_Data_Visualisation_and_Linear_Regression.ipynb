```python
import requests
import numpy as np
import matplotlib.pyplot as plt
import sklearn

url = 'https://api.github.com/search/repositories?q=language:python&sort=forks&per_page=50&page=8'
repositories = requests.get(url)
repo = repositories.json()
# print(repo) : imprime en une seule ligne
items = repo['items']
# forks = items[0] : premier de la liste
forks = [[repo['forks_count']] for repo in items]
print(forks)
stars = [repo['stargazers_count'] for repo in items]
print(stars)
```

```python
plt.scatter(forks, stars)
plt.xlabel('Forks')
plt.ylabel('Stars')
plt.show()

from sklearn.model_selection import train_test_split
forks_train, forks_test, stars_train, stars_test = train_test_split(forks, stars, test_size=1/5, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(forks_train, stars_train)
LinearRegression()

print(regressor.coef_)
print(regressor.intercept_)
```
