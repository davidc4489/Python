```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os
import urllib.request

def links(url, number):

    if number == 0:
        return
    
    response = requests.get (url)

    print(response.status_code)

    response.content

    soup = BeautifulSoup(response.content, 'html.parser')

    href = set()
    for link in soup.find_all('a', href=True):
        if link['href'].startswith('/wiki') and ':' not in link['href']:
            href.add(link['href'])
    print(len(href))

    for element in href:
        url = 'https://en.wikipedia.org' + element
        print(url)

    imgs = set()
    for link in soup.find_all('img', class_='thumbimage'):
        url = link['src']
        if url.startswith('//'):
            url = 'https:' + url
        elif url.startswith('/'):
            url = 'https://en.wikipedia.org' + url
        imgs.add(url)
    print(len(imgs))

    for element in imgs:
        print(element)

    base_directory = '/home/mefathim/Documents/Python/'

    dir_name = soup.find('title').text

    os.mkdir(base_directory+dir_name)

    directory = base_directory+dir_name
    
    def download_img(url, directory):
        response = requests.get(url)
        if response.status_code != 200:
            return
        x = 1
        for element in imgs:
            file_name = 'file'+ str(x) +'.jpg'
            myfile = open(directory+'/'+f'{file_name}', "wb") 
            myfile.write(urllib.request.urlopen(f'{element}').read())
            x += 1
            
    download_img(url,directory)

    for x in href:
            url = 'https://en.wikipedia.org' + x
            links(url, number-1)

url = input("please enter the url")
number = int(input("please enter number"))
links(url, number)

f = open('NASA2.jpg','wb')
f.write(urllib.request.urlopen('http://www.python.org/images/success/nasa.jpg').read())
f.close()
print("download successful")

myfile = open('name.png', "wb") 
myfile.write(urllib.request.urlopen('https://upload.wikimedia.org/wikipedia/commons/thumb/b/bd/Python_Powered.png/220px-Python_Powered.png').read())
myfile.close()
print("download successful")
```
