```python
# Example : Create a 2D list
matrix = [[x for x in range(1, 4)] for k in range (3)]
print(matrix)
# Output : [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
```

```python
# Example : Combine elements from 2 lists
numbers = [1,2,3]
letters = ['A', 'B', 'C']
combined = [(x,y) for x in numbers for y in letters]
print(combined)
# Output : [(1, 'A'), (1, 'B'), (1, 'C'), (2, 'A'), (2, 'B'), (2, 'C'), (3, 'A'), (3, 'B'), (3, 'C')]
```

```python
# Example : Create a set of squared numbers
squared_set = {x**2 for x in range(1, 6)}
print(squared_set)
# Output : {1, 4, 9, 16, 25}
```

```python
# Example : Create a dictionnary of numbers and their squares
squared_dict = {x: x**2 for x in range (1, 6)}
print(squared_dict)
# Output : {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```
