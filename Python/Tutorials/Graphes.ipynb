```python
# Matplotlib est une bibliothèque python qui dessine des graphiques. 
# Nul besoin de connaissances en interfaces graphiques pour créer un graphique dynamique
#   avec possibilité de zoom et de sauvegarde par l'utilisateur.
# Il est d'ailleurs possible de sauvegarder les graphiques en format matriciels comme le PNG, JPEG , etc.
#   et vectoriels comme le PDF et le SVG.
# Une grande communauté s'est formée autour de ce projet et donc la documentation et les exemples ne manquent pas.

import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
plt.ylabel('Label 1')
plt.show()

# Dans cet exemple on voit que par défaut il n'y a pas qu'un graphique, 
#   il y a aussi des outils pour exploiter ce graphique: 
#   il est en effet possible de zoomer, de modifier la taille du rendu et bien sur de l'enregistrer.
#
# Pour visualiser le gaphique en plein écran vous pouvez utiliser le raccurci Ctrl + F. 
# Il est possible d'afficher une grille en appuyant sur la touche g de votre clavier.
#
# Dans notre exemple nous avons renseigné uniquement les valeurs de l'abscisse, c'est à dire [1,2,3,4],
#   la lib a donc remplit la valeur des ordonnées d'elle-même comme si nous avions donné comme valeur
#   range(4) c'est à dire [0,1,2,3]
```

```python
# Il est possible d'associer une valeur x à une valeur y comme ceci:

plt.title("Danger de la vitesse")
plt.plot([50,100,150,200], [1,2,3,4])
plt.xlabel('Vitesse')
plt.ylabel('Temps')
plt.show()
```

```python
# axis(xmin, xmax, ymin, ymax)
# La courbe est juste mais je ne souhaite délimiter moi même les limites du graphique.
# Seul les vitesses entre 80 et 180 m'intéressent et le temps de 1 secondes à 10 secondes.
# Il est possible d'encadrer son graphique à l'aide de la méthode axis(xmin, xmax, ymin, ymax)

plt.plot([50,100,150,200], [1,2,3,4])
plt.xlabel('Vitesse')
plt.ylabel('Temps')
plt.axis([80, 180, 1, 10])
plt.show()
```

```python
# Il est possible de superposer les courbes:

plt.plot([50,100,150,200], [1,2,3,4])
plt.plot([50,100,150,200], [2,3,7,10])
plt.plot([50,100,150,200], [2,7,9,10])
plt.show()

# Vous pouvez utiliser cette syntaxe également:

plt.plot([50,100,150,200], [1,2,3,4], [50,100,150,200], [2,3,7,10], [50,100,150,200], [2,7,8,9])
plt.show()
```

```python
# Il est possible de changer la couleur et la forme des courbes:
# Courbes possibles: - -- -. :
# Couleurs possibles: b g r c m y k w

plt.plot([50,100,150,200], [1,2,3,4], "r--")
plt.plot([50,100,150,200], [2,3,7,10], "bs")
plt.plot([50,100,150,200], [2,7,9,10], "g^")
plt.show()

#Vous pouvez utiliser cette syntaxe également:

plt.plot([50,100,150,200], [1,2,3,4], "r--", [50,100,150,200], [2,3,7,10], "bs", [50,100,150,200], [2,7,9,10], "g^")
plt.show()

# Vous pouvez changer la largeur des courbes comme ceci:

plt.plot([50,100,150,200], [1,2,3,4], "r--", linewidth=5)
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=3)
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=10)
plt.show()
```

```python
# Il est possible de mettre plusieurs graphiques sur une même image:

plt.subplot(211)
plt.plot([50,100,150,200], [1,2,3,4], "r--", linewidth=5)
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=3)
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=10)
plt.xlabel('Vitesse')
plt.ylabel('Temps')
plt.axis([80, 180, 1, 10])

plt.subplot(212)
plt.plot([50,100,150,200], [1,2,3,15], "r--", linewidth=5)
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=3)
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=10)
plt.xlabel('Vitesse')
plt.ylabel('Temps')
plt.axis([80, 180, 1, 10])

plt.show()

# plt.subplot(lignes, colonnes, position)
plt.subplot(2,3,1)
plt.plot([1,2,3,4],[5,6,7,8])
plt.subplot(2,3,3)
plt.plot([2,3,4,5],[16,17,18,19])
plt.subplot(2,3,4)
plt.plot([3,4,5,6],[7,8,9,10])
plt.show()

plt.subplot(2,3,1)
plt.plot([1,2,3,4],[5,6,7,8])
plt.plot([2,3,4,5],[16,17,18,19])
plt.plot([3,4,5,6],[7,8,9,10])
plt.show()

```

```python
# Vous pouvez également ajouter une grille et des marqueurs:
# marqueurs possibles : o + . x * ^

plt.grid(True)
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=0.8, marker="*")
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=0.8, marker="+")
# Il est possible d'ajouter du texte en indiquant les coordonnées:
plt.text(150, 6.5, r'Danger')
plt.show()

# Vous pouvez également modifier la couleur de la grille

plt.grid(color='r')
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=0.8, marker="*")
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=0.8, marker="+")
plt.show()

# Vous pouvez également rajouter des grilles mineures
plt.minorticks_on()
plt.grid(which='minor',color='r')
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=0.8, marker="*")
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=0.8, marker="+")
plt.show()
```

```python
# Ajouter une légende pour les courbes

plt.grid(True)
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=0.8, marker="*", label="Trajet 1")
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=0.8, marker="+", label="Trajet 2")
plt.axis([80, 180, 1, 10])
plt.xlabel('Vitesse')
plt.ylabel('Temps')
plt.legend()    
plt.show()
```

```python
# Ajouter une flèche descriptive

plt.grid(True)
plt.plot([50,100,150,200], [2,3,7,10], "b", linewidth=0.8, marker="*")
plt.plot([50,100,150,200], [2,7,9,10], "g", linewidth=0.8, marker="+")
plt.axis([80, 180, 1, 10])
plt.annotate('Limite', xy=(150, 7), xytext=(165, 5.5), 
arrowprops={'facecolor':'black', 'shrink':0.05} )
plt.xlabel('Vitesse')
plt.ylabel('Temps')
plt.show()
```

```python
# pie / diagramme circulaire
# Un des graphiques les plus utilisés doit être le diagramme circulaire (ou camembert ) :

name = ['-18', '18-25', '25-50', '50+']
data = [5000, 26000, 21400, 12000]
explode=(0, 0.15, 0, 0)
plt.pie(data, explode=explode, labels=name, autopct='%1.1f%%', startangle=90, shadow=True)
plt.axis('equal')
plt.show()
# Les données explode indiquent les espaces entre les parts.
# Nous avons fait en sorte dans cet exemple de séparer la part 18-25 des autres part. 
# Le paramètre autopct indique le pourcentage réel par rapport aux données qu'on lui a fournit. 
# Enfin la méthode axis("equal") rend le diagramme parfaitement circulaire. 
# En son absence lorsque vous redimensionnez le graphique celui-ci s'adapte au conteneur ce qui peut lui donner une forme élliptique.
```

```python
# Scatter plot (nuage de points) :

x = [1, 2, 3, 4, 5]; 
y = [1, 2, 3, 4, 5]; 
plt.scatter(x, y)
plt.show()

# on peut tracer deux nuages de points simultanément sur le même graphe :

x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
plt.scatter(x, y1, c = 'red')
plt.scatter(x, y2, c = 'yellow')
plt.show()
```

```python
# graphe en 3D
import plotly.graph_objects as go
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
z = [1, 4, 9, 16, 25]
trace = go.Scatter3d( x = x, y = y, z = z,
                        mode = 'lines',
                        line = dict(color = 'blue', width = 2))
data = [trace]
layout = go.Layout (scene=dict(
                    xaxis_title='Y',
                    yaxis_title='X',
                    zaxis_title='Time',
                    ), title='Random Walk in 3D')
fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
# graphe en 3D
from mpl_toolkits import mplot3d
import numpy as np
ax = plt.axes(projection="3d")
plt.show()

ax = plt.axes(projection="3d")
z = np.linspace (0, 2*np.pi, 20)
x = np.cos(z)
y = np.sin(z)
ax.plot3D(x,y,z)
plt.show()
```

```python
# Pour créer un histogramme on utilise la méthode hist.
# On peut lui donner des données brutes et il s'occupera 
# de faire les calcules nécessaires à la présentation du graphique.
# On a prit ici l'exemple de 1000 tirage au sort (random) de valeur entre 0 et 150

# Pour créer un histogramme on utilise la méthode hist.
# On peut lui donner des données brutes et il s'occupera 
# de faire les calcules nécessaires à la présentation du graphique.
# On a prit ici l'exemple de 1000 tirage au sort (random) de valeur entre 0 et 150

valeur = [0, 1.2, 4.6, 7, 8.7, 11.4, 15.6]
plt.hist(valeur)
plt.show()


# Pour changer la repartition de l'histogramme, il faut utiliser le parametre bins
valeur = [0, 1.2, 4.6, 7, 8.7, 11.4, 15.6]
plt.hist(valeur, bins = 4)
plt.show()
```

```python
# numpy est une bibliothèque python qui sert aux calculs numeriques 
import numpy as np

x = np.linspace (0, 2*np.pi, 20) # point de depart du graphique [0], point d'arrivee [2*np.pi], nombre de points [20]
y = np.cos(x)
plt.plot(x,y)
plt.show()

x = np.linspace (0, 2*np.pi, 20)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x,y1, label="cos(x)")
plt.plot(x,y2, label="sin(x)")
plt.legend()
plt.show()
```
