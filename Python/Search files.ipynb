```python
def isin(txt,file):
    try:
        return txt in open(file,'r').read()
    except:
#         print(file)
        return False
        
    
def search_in_file(text, suffix='.ipynb'):
    roots = []
    i=0
    for root, dirs, files in os.walk('/home'):
        if suffix:
            files =  [file for file in files
                        if file.endswith(suffix)]
        match = [file for file in files
                 if isin(text,os.path.join(root,file))]
    #     res.extend([os.path.join(root,file) for file in ipynb])
        if match:
            roots.append(root)
            print(i, root)
            print('\t',*sorted(match),sep='\n\t')
            print('\n'*2)
            i+=1
    
    subprocess.call(f"nautilus '{roots[int(input('folder to open'))]}' --browser", shell=True)
```
