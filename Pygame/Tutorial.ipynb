```python
import pygame

pygame.init()
```

```python
screen = pygame.display.set_mode((400,400))

running = True

image = pygame.image.load("ball.jpg").convert()     # recuperer une image et comversion en format le mieux adapte

while running:
    for event in pygame.event.get():    # liste des evenements
        if event.type == pygame.QUIT:   # constante de la fermeture de la fenetre
            running = False

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                print("Left!")

            if event.key == pygame.K_RIGHT:
                print("Right!")

            if event.key == pygame.K_UP:
                print("Up!")

            if event.key == pygame.K_DOWN:
                print("Down!")

    screen.blit(image,(0,0))

    pygame.display.flip()

pygame.quit()   # Desinitialisation des elements de pygame
```

```python
screen = pygame.display.set_mode((400,400))

running = True

image = pygame.image.load("ball.jpg").convert()     # recuperer une image et comversion en format le mieux adapte

x = 0
y = 0

clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():    # liste des evenements
        if event.type == pygame.QUIT:   # constante de la fermeture de la fenetre
            running = False

    pressed = pygame.key.get_pressed()

    if pressed[pygame.K_LEFT]:
        x -= 1
    if pressed[pygame.K_RIGHT]:
        x += 1
    if pressed[pygame.K_UP]:
        y -= 1
    if pressed[pygame.K_DOWN]:
        y += 1      

    screen.fill((0,0,0))        # recolore en noir la surface d'affichage pour "effacer" les pixels precedents

    screen.blit(image,(x,y))

    pygame.display.flip()

    clock.tick(600)              # Pour que la boucle de jeu tourne a 60 frames par seconde [et pas 5000]

pygame.quit()   # Desinitialisation des elements de pygame
```
